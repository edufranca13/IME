1. leia um número inteiro positivo n e dois valores reais a e b, gere um histograma da lista resultante da função transformacaoEscala, grafique o histograma e salve a imagem na pasta de execução do programa com o nome hist_escala_translacao.png;

def transformacaoEscala(U, V, n, a, b): A função deve recebe uma duas listas, U e V, ambas de tamanho n, U com valores amostrados de uma distribuição uniforme, por exemplo, números entre 0 e 1, além de dois valores reais, a e b.
A sua função deve transformar a lista V de modo que, para uma posição genérica i de V o novo valor seja igual a: a + U[i]*(b-a+1).

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

def transformacaoEscala(U, V, n, a, b):
    for i in range(n):
        V[i] = a+U[i]*(b-a+1)
        
if __name__ == '__main__':

    n = int(input('Entre com numero inteiro positivo: '))
    a = float(input('Entre com a: '))
    b = float(input('Entre com b: '))
    
    U = np.random.uniform(0.0, 1.0, n)
    V = np.random.uniform(0.0, 1.0, n)
    
    transformacaoEscala(U,V, n, a, b)
    
    fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))
    ax0.hist(U, n)
    ax1.hist(V, n)
    ax0.grid()
    ax1.grid()
    plt.show()
```

Análise da função transformacaoEscala:
Como a função multiplica a lista original pelo fator (b-a+1), temos que o eixo dos intervalos é modificado, pois os valores se modificam todos na mesma proporção. A função também soma o fator a e, portanto, translada os valores no eixo. O padrão da distribuição do histograma não se altera, pois, como explicado, os valores foram todos alterados do mesmo modo. 

2. grafique o histograma da lista resultante da fun ̧ca ̃o somaVetores e salve a imagem na pasta de execuc ̧ ̃ao do programa com o nome hist soma.png.

def somaVetores(U1, U2, U, n): A função deve receber duas listas, U1 e U2, ambas de tamanho n, com valores amostrados uniformemente. Além delas, uma lista U de mesmo tamanho, preenchida com zeros (pode ser qualquer número, na verdade).

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

def somaVetores(U1, U2, U, n):
    for i in range(n):
        U[i] = U1[i]+U2[i]
   
if __name__ == '__main__':

    n = int(input('Entre com numero inteiro positivo: '))

    U = [0]*n
    U1 = np.random.uniform(0.0, 1.0, n)
    U2 = np.random.uniform(0.0, 1.0, n)
    
    somaVetores(U1, U2, U, n)
    
    fig, (ax0,ax1,ax2) = plt.subplots(ncols=3, figsize=(8, 4))
    ax0.hist(U1, n)
    ax0.grid()
    ax1.hist(U2, n)
    ax1.grid()
    ax2.hist(U, n)
    ax2.grid()    
    plt.show()
```

Como pode ser observado a soma de duas distribuições uniformes leva a uma distribuição normal (em torno de 1, no exemplo). 



3. grafique o histograma da lista resultante da função raizVetor e salve a imagem na pasta de execução do programa com o nome hist_raiz.png.

def raizVetor(U,n): A função deve recebe uma lista, U, de tamanho n, U com valores amostrados uniformemente, como anteriormente.
A sua função deve transformar a lista U de modo que, para uma posição genérica i de U, o novo valor seja igual a: raiz(U[i]).


```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

def raizVetor(U,n):
    U1 = list()
    for i in range(n):
        U1.append(sqrt(U[i]))
    return U1

if __name__ == '__main__':

    n = int(input('Entre com numero inteiro positivo: '))
    
    U = np.random.uniform(0.0, 1.0, n)
    U1 = raizVetor(U, n)
    
    fig, (ax0,ax1) = plt.subplots(ncols=2, figsize=(8, 4))
    ax0.hist(U,n)
    ax0.grid()
    ax1.hist(U1, n)
    ax1.grid()
    plt.show()
```

Como se pode observar a função raizVetor modifica a distribuição normal para algo como uma distribuição de mínimos. Isso é esperado, visto que a raiz de um número entre 0 e 1 é maior que o próprio o número.

4. grafique o histograma da lista resultante da função normalTransform para as listas Z0 e Z1 e salve a imagem na pasta de execução do programa com os nomes hist_normal_Z0.png e hist_normal_Z1.png, respectivamente.

def normalTransform(U1, U2, Z1, Z2, n): A função deve receber duas listas, U1 e U2, ambas de tamanho n, com valores amostrados uniformemente no intervalo (0, 1). Além delas, duas listas, Z1 e Z2 de mesmo tamanho das primeiras, preenchida com zeros.
A sua função deve transformar Z1 e Z2 de forma que, para uma posição genérica i, elas sejam iguais a:
Z0[i] = cos(2πU2[i]) ∗  −2 ln U1[i] Z1[i] = sin(2πU2[i]) ∗  −2 ln U1[i]
Obs.: A transformação acima é conhecida como Box-Muller. Para quem tiver curiosidade, veja a página da Wikipedia: Wikipedia - Box-Muller Transform.

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from math import cos, sin, sqrt, log, pi

def normalTransform(U1, U2, Z0, Z1, n):
    for i in range(n):
        Z0[i] = cos(2*pi*U2[i])*sqrt(-2*log(U1[i]))
        Z1[i] = sin(2*pi*U2[i])*sqrt(-2*log(U1[i]))
        
if __name__ == '__main__':

    n = int(input('Entre com numero inteiro positivo: '))
    
    U1 = np.random.uniform(0.0, 1.0, n)
    U2 = np.random.uniform(0.0, 1.0, n)
    Z0 = [0]*n
    Z1 = [0]*n
    
    normalTransform(U1, U2, Z0, Z1, n)
    
    fig, (ax0,ax1) = plt.subplots(ncols=2, figsize=(8, 4))
    ax0.hist(U1, n)
    ax1.hist(U2, n)
    ax0.grid()
    ax1.grid()
    plt.show()
    
    fig, (ax0,ax1) = plt.subplots(ncols=2, figsize=(8, 4))
    ax0.hist(Z0, n)
    ax1.hist(Z1, n)
    ax0.grid()
    ax1.grid()
    plt.show()
    
    
```

Como se pode observar a função faz o esperado transformando distribuições uniformes em distribuições normais. 

Análise da distribuição das cordas geradas no EP2

Para cada tipo de distribuição será usado os três métodos de geração de pontos para geração de cordas.
O método 1 gera pares de pontos aleatorios distribuidos uniformemente sobre a circuferência. 
O método 2 gera pontos aleatorios distribuidos uniformemente sobre um raio da circuferência. Esses pontos representam o ponto médio das cordas.
O método 3 gera pontos aleatorios distribuidos uniformemente dentro do circulo. Esses pontos representam o ponto médio das cordas.

Análise da distribuição dos pontos médios das cordas em área - def distribuicaoArea(r,listaCordas):
Essa função deve receber o tamanho (r) do raio e uma lista de tuplas contendo as coordenadas dos pontos extremos das n cordas.
Utilizando esses pontos, sua função deve:
- Determinar a lista dos n pontos médios de cada corda a partir de seus pontos extremos; 
- Fazer uma partição do círculo em 8 regiões de mesma área;
- Categorizar os pontos médios das cordas de acordo com a região a qual o ponto pertence. Para isso, pode-se criar uma lista que contém os nomes das categorias dos n pontos médios;
- Em um gráfico, esboçar o círculo de raio r e os n pontos médios categorizados por cor;
- Ao lado, na mesma imagem, esboçar o histograma que mostra a frequência (quantidade de pontos pertencentes) de cada categoria;
- Salvar a imagem em formato png na mesma pasta de execução do programa com o nome ”distribuicao_area.png”.



```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from random import random
from math import sqrt, degrees, acos, fabs

def points_on_circle(r):
    return (r, 360*random())

def points_in_circle(r):
    return (sqrt(random())*r, 360*random())

def determinaPontosExtremos(listaMedios):
    ponto_extremos = list()
    for i in range(len(listaMedios)):
        a1 = listaMedios[i][1] - degrees(acos(float(listaMedios[i][0])/r))
        if a1 > 360:
            while a1 > 360:
                a1 -= 360
        if a1 < 0:
            while a1 < 0:
                a1 += 360
        a2 = listaMedios[i][1] + degrees(acos(float(listaMedios[i][0])/r))
        if a2 > 360:
            while a2 > 360:
                a2 -= 360
        if a2 < 0:
            while a2 < 360:
                a2 += 360
        r1,r2 = float(r), float(r)
        ponto_extremos.append(((r1,a1), (r2, a2)))
    return ponto_extremos


def determinaPontosMedios(listaExtremos):

    pontos_medios = list()

    for i in range(len(listaExtremos)):
        if listaExtremos[i][0][1] > listaExtremos[i][1][1]:
            alfa = listaExtremos[i][1][1] + (listaExtremos[i][0][1] - listaExtremos[i][1][1])/2
            rm = np.cos(np.radians(alfa - listaExtremos[i][1][1]))*r
        else:
            alfa = listaExtremos[i][0][1] + (listaExtremos[i][1][1] - listaExtremos[i][0][1])/2
            rm = np.cos(np.radians(alfa - listaExtremos[i][0][1]))*r
        pontos_medios.append((rm, alfa))

    return pontos_medios

def listaCordasM1(r, n):
    return [(points_on_circle(r), points_on_circle(r)) for i in range(n)]

def listaCordasM2(r,n):
    pontos_medios = [(r*random(), 360*random()) for i in range(n)]
    return determinaPontosExtremos(pontos_medios)

def listaCordasM3(r,n):
    pontos_medios = [(points_in_circle(r)) for i in range(n)]
    return determinaPontosExtremos(pontos_medios)

def extract(lista, a):
    lista2 = list()
    for i in range(len(lista)):
        lista2.append(lista[i][0][a])
        lista2.append(lista[i][1][a])
    return lista2

def extract_m(lista, a):
    return [lista[i][a] for i in range(len(lista))]

def find_key(d, value):
    for keys in d:
        for v in d[keys]:
            if v == value:
                s = str(keys)
                return s
    return None

def classify_a(lista):
    lista2 = lista[:]
    lista2.sort()

    d = dict()

    categories = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'white']

    j = 0
    l = 0
    for j, cat in enumerate(categories):
        for i in range(l, len(lista2)):
            if lista2[i] <= 45*(j+1):
                if cat in d:
                    d[cat].append(lista2[i])
                else:
                    d[cat] = [lista2[i]]
                l += 1
    return d

def classify_r(lista):

    lista2 = [fabs(lista[i]) for i in range(len(lista))]
    lista2.sort()

    d = dict()

    categories = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'white']

    j = 0
    l = 0
    for j, cat in enumerate(categories):
        for i in range(l, len(lista2)):
            if lista2[i] <= r*(j+1)/8:
                if cat in d:
                    d[cat].append(lista2[i])
                else:
                    d[cat] = [lista2[i]]
                l += 1
    return d

def classify_r2(lista):

    lista2 = [fabs(lista[i]) for i in range(len(lista))]
    lista2.sort()

    d = dict()

    categories = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'white']

    j = 0
    l = 0
    for j, cat in enumerate(categories):
        for i in range(l, len(lista2)):
            if lista2[i] <= r*sqrt((j+1)/8):
                if cat in d:
                    d[cat].append(lista2[i])
                else:
                    d[cat] = [lista2[i]]
                l += 1
    return d

def distribuicaoBorda(listaCordas):

    angles = extract(listaCordas, 1)

    d = classify_a(angles)

    fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))
    ax0.plot(r*np.cos(np.arange(0, 2*np.pi, 0.01)), r*np.sin(np.arange(0, 2*np.pi, 0.01)), '-b')

    for a in angles:
        ax0.plot([r*np.cos(np.pi*a/180)], [r*np.sin(np.pi*a/180)], color=find_key(d, a), marker='o')
        ax0.grid()
 
    ax1.hist(angles, 8, range = [0, 360])
    ax1.grid()

    plt.show()
    
def distribuicaoRadial(listaCordas):

    pontos_medios = determinaPontosMedios(listaCordas)

    angles = extract_m(pontos_medios, 1)
    radius = extract_m(pontos_medios, 0)

    d = classify_r(radius)

    fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))
    ax0.plot(r*np.cos(np.arange(0, 2*np.pi, 0.01)), r*np.sin(np.arange(0, 2*np.pi, 0.01)), '-b')

    for ra, a in zip(radius, angles):
        ax0.plot([ra*np.cos(np.pi*a/180)], [ra*np.sin(np.pi*a/180)], color=find_key(d, np.fabs(ra)), marker='o')
    ax0.grid()

    rp = [np.fabs(radius[i]) for i in range(len(radius))]
    ax1.hist(rp, 8, range = [0, r])
    ax1.grid()

    plt.show()
    
def distribuicaoArea(listaCordas):

    pontos_medios = determinaPontosMedios(listaCordas)

    angles = extract_m(pontos_medios, 1)
    radius = extract_m(pontos_medios, 0)

    d = classify_r2(radius)

    fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))
    ax0.plot(r*np.cos(np.arange(0, 2*np.pi, 0.01)), r*np.sin(np.arange(0, 2*np.pi, 0.01)), '-b')

    for ra, a in zip(radius, angles):
        ax0.plot([ra*np.cos(np.pi*a/180)], [ra*np.sin(np.pi*a/180)], color=find_key(d, np.fabs(ra)), marker='o')
    ax0.grid()

    rp = [np.fabs(radius[i]) for i in range(len(radius))]
    ax1.hist(rp, [r*np.sqrt((j+1)/8) for j in range(8)], range = [0, r])

    ax1.grid()

    plt.show()

def distribuicaoCordas(listaCordas):

    fig, ax0 = plt.subplots(ncols=1, figsize=(8, 4))
    ax0.plot(r*np.cos(np.arange(0, 2*np.pi, 0.01)), r*np.sin(np.arange(0, 2*np.pi, 0.01)), '-b')

    xzes = []
    yzes = []
    c = 0

    for par in listaCordas:
        temp1 = []
        temp2 = []
        for ponto in par:
            temp1.append(ponto[0]*np.cos(np.pi*ponto[1]/180))
            temp2.append(ponto[0]*np.sin(np.pi*ponto[1]/180))
        xzes.append(temp1)
        yzes.append(temp2)

    for x, y in zip(xzes, yzes):
        ax0.plot(x, y, 'r-')

    ax0.grid()

    plt.show()
    
if __name__ == '__main__':
    
    r = float(input("Entre com o r: "))
    n = int(input("Entre com o n: "))
    
    lista1 = listaCordasM1(r, n)
    lista2 = listaCordasM2(r, n)
    lista3 = listaCordasM3(r, n)

    distribuicaoBorda(lista1)
    distribuicaoRadial(lista1)
    distribuicaoArea(lista1)
    distribuicaoCordas(lista1)

    distribuicaoBorda(lista2)
    distribuicaoRadial(lista2)
    distribuicaoArea(lista2)
    distribuicaoCordas(lista2)

    distribuicaoBorda(lista3)
    distribuicaoRadial(lista3)
    distribuicaoArea(lista3)
    distribuicaoCordas(lista3)
```
